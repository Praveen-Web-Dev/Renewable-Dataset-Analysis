```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
# Load Excel with multiple sheets
meta_file = pd.ExcelFile('metadata.xlsx')

# View available sheets (optional)
print("Available sheets:", meta_file.sheet_names)

# Load the specific metadata sheet
df_meta = meta_file.parse('List of economies')

# Preview
df_meta.head()

# Load your main dataset with country-year energy/emissions data
df_main = pd.read_csv('Dataset.csv')

# Preview
df_main.head()

# Strip whitespace
df_main['Entity'] = df_main['Entity'].str.strip()
df_meta['Economy'] = df_meta['Economy'].str.strip()


# Fix known mismatches between Entity and Economy naming
country_fix_map = {
    'Bahamas': 'Bahamas, The',
    'Congo': 'Congo, Rep.',
    'Egypt': 'Egypt, Arab Rep.',
    'French Guiana': 'France',
    'Gambia': 'Gambia, The',
    'Kyrgyzstan': 'Kyrgyz Republic',
    'Puerto Rico': 'Puerto Rico',
    'Saint Kitts and Nevis': 'St. Kitts and Nevis',
    'Saint Lucia': 'St. Lucia',
    'Saint Vincent and the Grenadines': 'St. Vincent and the Grenadines',
    'Sao Tome and Principe': 'São Tomé and Príncipe',
    'Slovakia': 'Slovak Republic',
    'Turkey': 'Türkiye',
    'Yemen': 'Yemen, Rep.'
}

# Apply the mapping
df_main['Entity'] = df_main['Entity'].replace(country_fix_map)

# Merge on country name
merged_df = df_main.merge(
    df_meta[['Economy', 'Region', 'Income group', 'Lending category']],
    left_on='Entity',
    right_on='Economy',
    how='left'
)

# Drop redundant column
merged_df.drop(columns='Economy', inplace=True)


# Add metadata for Puerto Rico manually
merged_df.loc[merged_df['Entity'] == 'Puerto Rico', 'Region'] = 'Latin America & Caribbean'
merged_df.loc[merged_df['Entity'] == 'Puerto Rico', 'Income group'] = 'High income'
merged_df.loc[merged_df['Entity'] == 'Puerto Rico', 'Lending category'] = 'Not classified'

# Preview
merged_df.head()


# Check if there are any countries still not matched
unmatched = merged_df[merged_df['Region'].isnull()]['Entity'].unique()
print("Still unmatched:", unmatched)


# Save the unified dataset
merged_df.to_csv('Unified_Cleaned_Dataset.csv', index=False)

# Download the file
files.download('Unified_Cleaned_Dataset.csv')
```
