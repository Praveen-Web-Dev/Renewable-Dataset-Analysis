```python
# Step 0 â€“ Upload and load dataset in Google Colab
from google.colab import files
uploaded = files.upload()

import pandas as pd
df = pd.read_csv('cleaned_dataset.csv')  # Replace filename if different
df.head()

import statsmodels.api as sm
import numpy as np

# Prepare data
df1 = df[['value_co2_emissions_kt_by_country',
          'renewable_energy_share_in_the_total_final_energy_consumption_(%)']].dropna()
df1['log_co2'] = np.log(df1['value_co2_emissions_kt_by_country'] + 1)

X = sm.add_constant(df1['renewable_energy_share_in_the_total_final_energy_consumption_(%)'])
y = df1['log_co2']

# Run regression
model1 = sm.OLS(y, X).fit()
model1.summary()
```
