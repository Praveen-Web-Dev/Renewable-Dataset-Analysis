```python
# Step 1: Upload your dataset
from google.colab import files
uploaded = files.upload()

# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Load dataset
df = pd.read_csv('cleaned_dataset.csv')  # Make sure filename matches

# Step 4: Filter for India and China with non-null values
df_stack = df[df['entity'].isin(['India', 'China'])][[
    'entity', 'year',
    'electricity_from_fossil_fuels_(twh)',
    'electricity_from_renewables_(twh)'
]].dropna()

# Step 5: Generate plots for each country
for country in ['India', 'China']:
    country_df = df_stack[df_stack['entity'] == country]

    plt.figure(figsize=(10, 6))
    plt.plot(country_df['year'], country_df['electricity_from_fossil_fuels_(twh)'],
             label='Fossil Fuels', linestyle='--', marker='o')
    plt.plot(country_df['year'], country_df['electricity_from_renewables_(twh)'],
             label='Renewables', linestyle='-', marker='s')
    plt.title(f"Electricity Generation Trends in {country} (2000–2020)")
    plt.xlabel("Year")
    plt.ylabel("Electricity Generation (TWh)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
