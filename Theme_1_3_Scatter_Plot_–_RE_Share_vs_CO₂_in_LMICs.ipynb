```python
# Upload your cleaned dataset
from google.colab import files
uploaded = files.upload()

# Load the dataset (change filename if needed)
import pandas as pd
import matplotlib.pyplot as plt # Import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_dataset.csv')  # Replace with actual uploaded filename
df.head()



# Filter for LMICs
df_lmic = df[df['income_group'].isin(['Low income', 'Lower middle income'])]

# Clean data
df_lmic_clean = df_lmic[[
    'renewable_energy_share_in_the_total_final_energy_consumption_(%)',
    'value_co2_emissions_kt_by_country'
]].dropna()

# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_lmic_clean,
    x='renewable_energy_share_in_the_total_final_energy_consumption_(%)',
    y='value_co2_emissions_kt_by_country'
)
plt.title("Renewable Share vs CO₂ Emissions in LMICs")
plt.xlabel("Renewable Energy Share (%)")
plt.ylabel("CO₂ Emissions (kt)")
plt.grid(True)
plt.show()
```
