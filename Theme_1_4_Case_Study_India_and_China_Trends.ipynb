```python
# Upload your cleaned dataset
from google.colab import files
uploaded = files.upload()

# Load the dataset (change filename if needed)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned_dataset.csv')  # Replace with actual uploaded filename
df.head()

# Filter for India and China
df_ic = df[df['entity'].isin(['India', 'China'])]

df_ic_clean = df_ic[['entity', 'year',
                     'renewable_energy_share_in_the_total_final_energy_consumption_(%)',
                     'value_co2_emissions_kt_by_country']].dropna()

# Plot CO emissions
plt.figure(figsize=(10, 5))
for country in df_ic_clean['entity'].unique():
    subset = df_ic_clean[df_ic_clean['entity'] == country]
    plt.plot(subset['year'], subset['value_co2_emissions_kt_by_country'], label=country)
plt.title("CO Emissions: India vs China (20002020)")
plt.xlabel("Year")
plt.ylabel("CO Emissions (kt)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Renewable Share
plt.figure(figsize=(10, 5))
for country in df_ic_clean['entity'].unique():
    subset = df_ic_clean[df_ic_clean['entity'] == country]
    plt.plot(subset['year'], subset['renewable_energy_share_in_the_total_final_energy_consumption_(%)'], label=country)
plt.title("Renewable Energy Share: India vs China (20002020)")
plt.xlabel("Year")
plt.ylabel("Renewable Share (%)")
plt.legend()
plt.grid(True)
plt.show()
```
