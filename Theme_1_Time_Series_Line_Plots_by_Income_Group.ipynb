```python
# Upload your cleaned dataset
from google.colab import files
uploaded = files.upload()

# Load the dataset (change filename if needed)
import pandas as pd

df = pd.read_csv('cleaned_dataset.csv')  # Replace with actual uploaded filename
df.head()

import matplotlib.pyplot as plt

# Clean time-series data
df_time = df[['year', 'income_group',
              'value_co2_emissions_kt_by_country',
              'renewable_energy_share_in_the_total_final_energy_consumption_(%)']].dropna()

grouped_time = df_time.groupby(['year', 'income_group']).mean().reset_index()

# Plot CO₂ emissions over time
plt.figure(figsize=(10, 5))
for group in grouped_time['income_group'].unique():
    subset = grouped_time[grouped_time['income_group'] == group]
    plt.plot(subset['year'], subset['value_co2_emissions_kt_by_country'], label=group)
plt.title("CO₂ Emissions by Income Group (2000–2020)")
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (kt)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Renewable Share over time
plt.figure(figsize=(10, 5))
for group in grouped_time['income_group'].unique():
    subset = grouped_time[grouped_time['income_group'] == group]
    plt.plot(subset['year'], subset['renewable_energy_share_in_the_total_final_energy_consumption_(%)'], label=group)
plt.title("Renewable Energy Share by Income Group (2000–2020)")
plt.xlabel("Year")
plt.ylabel("Renewable Share (%)")
plt.legend()
plt.grid(True)
plt.show()
```
